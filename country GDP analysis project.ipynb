#  intro to seaborn --> statistic visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize']=6,2


import warnings
warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
df['InternetUsers']
```

```python
# distributions:

vis1 = sns.distplot(df["InternetUsers"])              # univariate statistics---> plot using one variable
plt.show(vis1)
```

```python
vis1= sns.displot(df["InternetUsers"])
plt.show(vis1)
```

```python
vis1 = sns.distplot(df["InternetUsers"], bins=10)
plt.show(vis1)
```

```python
vis2 = sns.boxplot(data=df, x="IncomeGroup", y="BirthRate")   # bi-variate analysis--> using 2 variable

# outlier will not change are data point which different from  data ======
# outlier:  in statistics outlier is the datapoint which is very far from datapoints
# how to detect outlier: using graphs(boxplot)

plt.show(vis2)
```

```python
vis3 = sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False)
plt.show(vis3)
```

```python
vis4=sns.lmplot(data=df,x='InternetUsers', y='BirthRate') #lm---> linear model
plt.show(vis4)
```

```python
vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',
                fit_reg=False,hue='IncomeGroup')
plt.show(vis5)
```

```python
vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',
                fit_reg=False,hue='IncomeGroup',size=10)
plt.show(vis5)
```

```python
vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',
                fit_reg=False,hue='IncomeGroup',height=10)
plt.show(vis5)
```
