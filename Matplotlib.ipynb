```python
%matplotlib inline

x1=np.linspace(0,10,100)
fig=plt.figure()

plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
```

```python
# histogram

data1=np.random.randn(1000)
plt.hist(data1)
```

```python
plt.show()
```

```python
# bar cahart

data2=[5.,25.,50.,20.,]
plt.bar(range(len(data2)),data2)
plt.show()
```

```python
# histogram bar chart:

data2=[5. ,25. ,50. ,20.]
plt.barh(range(len(data2)),data2)
plt.show()
```

```python
# error bar chart:

x9=np.arange(0,4,0.2)
y9=np.exp(-x9)
e1=0.1*np.abs(np.random.randn(len(y9)))
plt.errorbar(x9,y9,yerr=e1,fmt='.-')
plt.show()
```

```python
# stacked bar chart:

A=[15., 30., 45., 22.]
B=[15., 25., 50., 20.]
z2=range(4)
plt.bar(z2,A,color='b')
plt.bar(z2,B, color='r', bottom=A)
plt.show()
```

```python
# pie chart:

plt.figure(figsize=(7,7))
x10=[5,25,20,20]
labels=['computer', 'electronics','mechanical','chemical']
plt.pie(x10,labels=labels)
plt.show()
```

```python
# boxplot

data3 = np.random.randn(100)
plt.boxplot(data3)
plt.show()
```

```python
# area chart:

x12=range(1,6)
y12=[1,4,6,8,4]

plt.fill_between(x12,y12)
plt.show()
```

```python
# contour plot:

matrix1=np.random.rand(10,20)
cp=plt.contour(matrix1)
plt.show()
```

```python
# adding a legend:

x15=np.arange(1,5)
fig,ax=plt.subplots()
ax.plot(x15,x15*1.5)
ax.plot(x15,x15*3.0)
ax.plot(x15,x15/3.0)

ax.legend(['Normal', 'Fast', 'Slow'])
```

```python
# control colors:

x16 = np.arange(1, 5)
plt.plot(x16, 'r')
plt.plot(x16+1, 'g')
plt.plot(x16+2, 'b')
plt.show()
```
